### 4.5- Features Analysis

In this part, we especially take a look at the correlations.

```python
# Correlation matrix
mat_corr = data24.corr()
```

```python
mat_corr
```

Which variables are correlated to our target variable?

```python
th_corr = 0.5
for c in mat_corr.columns:
    if mat_corr["Hb_value_baseline"][c]>th_corr:
        print(c)
        print(mat_corr["Hb_value_baseline"][c])
        print("\n")
```

Correlations between features

```python
th_corr_min = 0.7
th_corr_max = 1
list_couples = []
list_features = []
for c in mat_corr.columns:
    for b in mat_corr.columns:
        if (np.abs(mat_corr[b][c])>th_corr_min) and (np.abs(mat_corr[b][c])<th_corr_max) and (b!=c) and not ((b,c) in list_couples) and b.split("_")[0]!=c.split("_")[0]:
            print(b)
            print(c)
            print(mat_corr[b][c])
            list_couples.append((b,c))
            list_couples.append((c,b))
            list_features.append(b)
            list_features.append(c)
            print("\n")
set_features = set(list_features)
```

```python
Var_Corr = data24[list(set_features)].corr()
sns.heatmap(Var_Corr, xticklabels=Var_Corr.columns, yticklabels=Var_Corr.columns, annot=True)
```
